```python
from pathlib import Path
from wordcounter import read_lyrics, clean, wordcounter
import matplotlib.pyplot as plt



DATA_DIR = Path.cwd().parent / 'data'
pesma1 = read_lyrics(DATA_DIR / "placeholder1.txt")
if pesma1: #MOŽE I BEZ OVOG IF-a, STAVIO SAM ZBOG SIGURNOSTI AKO READ_LYRICS NE VRATI NIŠTA
    reci1, frekvencije1 = wordcounter(clean(pesma1))
    fig, ax = plt.subplots(figsize=(15, 5))
    ax.bar(reci1, frekvencije1, color=['red', 'green', 'blue', 'orange'], edgecolor='black')
    ax.set_title('Broj reči u pesmi')
    ax.set_xlabel('Reči')
    ax.set_ylabel('Broj pojavljivanja')
    ax.tick_params(axis='x', rotation=90)
    plt.show()

```
